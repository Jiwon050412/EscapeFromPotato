```python
import gdown
gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv',quiet = False)
import matplotlib.pyplot as plt
import pandas as pd

ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
```

```python
plt.scatter(ns_book7['번호'], ns_book7['대출건수']) #(x좌표, y좌표)
plt.show()
```

```python
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
plt.hist([0,3,5,6,7,7,9,13], bins=5)
plt.show()
```

```python
plt.hist(ns_book7['대출건수']) #매개변수 log = True 로 지정해도 로그스케일 가능
plt.yscale('log')
plt.show()
```

```python
plt.boxplot(ns_book7[['대출건수','도서권수']])
plt.yscale('log')
plt.show()
```

```python
#수평으로 그리기
plt.boxplot(ns_book7[['대출건수','도서권수']], vert=False) #vert 매개변수를 통해 수평으로 그림
plt.xscale('log')
plt.show()
```

```python
#수염 길이 조정하기
plt.boxplot(ns_book7[['대출건수','도서권수']], whis=10)#whis 함수를 통해 수염길이 조절 10배 범위 안까지 수염 그리기 
plt.yscale('log')
plt.show()
```

```python
#확인문제 5번
selected_rows = (ns_book7['발행년도'] >= 1980) & (ns_book7['발행년도'] <= 2022)
plt.hist(ns_book7.loc[selected_rows, '발행년도'])
plt.show()
```

```python
plt.boxplot(ns_book7.loc[selected_rows,'발행년도'])
plt.show()
```
